```python
import pandas as pd
import numpy as np
```

```python
anno = pd.read_excel('final-rm_dup.xlsx')
```

```python
anno
```

```python
cur_d = None
dis_dict = {}
for index, row in anno.iterrows():
    d = "{}:{}".format(row.gene, row.condition_name)
    if d not in list(dis_dict.keys()):
        cur_d = d
        dis_dict[d] = [row.population, row.ethnic_group]
    else:
        dis_dict[d].extend([row.population, row.ethnic_group])
    if d == "ASPA:Canavan Disease":
        
        print(dis_dict[d])
```

```python
cols = ['Gene', 'Disease', 'Ethnicity']
lst = []
for k,v in dis_dict.items():
    s = set(v)
    try:
        s.remove('unknown')
        s.remove(np.nan)
        s.remove('unknown')
    except:
        pass
    st = ', '.join(list(s))
#     gen_mut = ', '.join(k.split(':'))
    lst.append([k.split(':')[0], k.split(':')[1], st])
df = pd.DataFrame(lst, columns=cols)
df.to_csv('anno_ethnicity.csv', index=False)
print(df)
#     print(gen_mut, st)
# print(len(dis_dict))
```
