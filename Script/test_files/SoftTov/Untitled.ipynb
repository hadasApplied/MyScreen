```python
cnv = pd.read_csv('cnv.anno.txt', sep='\t')
cnv = cnv.rename(columns={'Annotation1': 'Custom Annotation', 'Annotation2': 'Custom Annotation 2', 'Annotation3': 'Custom Annotation 3', 'Annotation4': 'Custom Annotation 4'})
cnv.columns
```

```python
UniqueSamples = all_geno.Sample.unique()
dfDict = {elem : pd.DataFrame for elem in UniqueSamples}
for key in dfDict.keys():
    dfDict[key] = all_geno[:][all_geno.Sample == key]
for k,v in dfDict.items():
        dfDict[k].reset_index(inplace=True)
        df = dfDict[k][dfDict[k]['AGID'].isin(list(pisces.agid.unique()))]
        print(len(df))
        print(len(pisces))
        #df.to_csv('test.csv')
        break
```

```python
 all_geno_bedouin = all_geno[all_geno['AGID'].isin(list(pisces.agid.unique()))]
```

```python
complete = pd.DataFrame(columns=list(all_geno.columns))
print(complete)
UniqueSamples = all_geno_bedouin.Sample.unique()
dfDict = {elem : pd.DataFrame for elem in UniqueSamples}
for key in dfDict.keys():
    dfDict[key] = all_geno_bedouin[:][all_geno_bedouin.Sample == key]
for k,v in dfDict.items():
        print(len(dfDict[k]))
        dfDict[k].reset_index(inplace=True)
        cnv['Sample'] = dfDict[k].loc[0,'Sample']
        dfDict[k] = dfDict[k].merge(cnv, on=['Sample', 'AGID', 'Custom Annotation', 'Custom Annotation 2', 'Custom Annotation 3',
       'Custom Annotation 4', 'gdna', 'Clalit Disease Makat',
       'Clalit Mutation Makat', 'Gene', 'Chr'], how='outer').reset_index()
        #print(dfDict[k])
        #df.to_csv('test.csv')
        complete = complete.append(dfDict[k], ignore_index = True) 
        #break
complete.to_csv('full.csv')
```
